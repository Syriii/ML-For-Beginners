```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
# 你可以看到这个数据有 442 个项目，组成了 10 个元素的数组：
# 10个元素指的是10个特征变量，比如age、bmi、bp等
# 但是数据看起来是已经经过归一化处理
# 数据归一化：糖尿病数据集的特征数据已经过归一化处理，使得特征值均值为0，方差为1。
# X：Each of these 10 feature variables have been mean centered and scaled by the standard deviation times the square root of n_samples (i.e. the sum of squares of each column totals 1).
# y：Column 11 is a quantitative measure of disease progression one year after baseline
```

```python
X = X[:, np.newaxis, 2]
print(X)

# np.newaxis用于增加数组的维度，使结果成为二维数组。
# 选择第2列数据作为新的特征矩阵X。
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```
